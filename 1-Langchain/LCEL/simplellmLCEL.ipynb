```python
import os
from dotenv import load_dotenv
load_dotenv()

import openai
open.api_keys = os.getenv("OPENAI_API_KEY")

groq_api_key = os.getenv("GROQ_API_KEY")
groq_api_key
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="gemma2-9b-it",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

messages = [
    SystemMessage(content="Traslate the followig from English to French"),
    HumanMessage(content="Hello, how are you?")
]

result = model.invoke(messages)
result
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
parser.invoke(result)
```

```python
# LCEL 
chain = model|parser
chain.invoke(messages)
```

```python
# Prompt template

from langchain_core.prompts import ChatPromptTemplate

generic_template = "Translate the following into {language}:"

promt = ChatPromptTemplate.from_messages(
    [
        ("system",generic_template),("user","{text}")
    ]
)

result = promt.invoke({"language":"French","text":"Hello"})
result
```

```python
result.messages
```

```python
# chain components together using LCEL

chain = promt|model|parser

chain.invoke({"language":"French","text":"Hello, how are you"})
```
